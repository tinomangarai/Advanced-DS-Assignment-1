## Feature Engineering 

```python
# Creating a function to categorize BMI into categories
def categorize_bmi(bmi):
    if bmi < 18.5:
        return 'Underweight'
    elif 18.5 <= bmi < 25:
        return 'Normal Weight'
    elif 25 <= bmi < 30:
        return 'Overweight'
    else:
        return 'Obese'

# Applying the function to create the new feature 'BMI_Category'
df['BMI_Category'] = df['BMI'].apply(categorize_bmi)

# Printing the updated dataset
df.head(5)
```

Categorizing the continuous variable 'BMI' into categories (Underweight, Normal Weight, Overweight, Obese) simplifies the analysis by providing a clear distinction between different BMI ranges. This can help in identifying patterns and relationships with other variables more easily. The new feature 'BMI_Category' provides a more interpretable representation of the individuals' BMI status. 

```python
# Creating a function to categorize blood pressure into categories
def categorize_bp(bp):
    if bp < 120:
        return 'Normal'
    elif 120 <= bp < 130:
        return 'Elevated'
    elif 130 <= bp < 140:
        return 'Stage 1 Hypertension'
    else:
        return 'Stage 2 Hypertension'

# Applying the function to create the new feature 'BP_Category'
df['BP_Category'] = df['BloodPressure'].apply(categorize_bp)
df.head(5)
```

The new features 'BMI_Category' and 'BP_Category' can be used as inputs in machine learning models to assess their importance in predicting the 'Outcome' variable. It may provide a more meaningful representation of the relationship between BMI and diabetes compared to using the raw numerical 'BMI' value.

## Task 3: Statistical Analysis 

```python
# Descriptive statistics 
stats = df.drop('Outcome', axis = 1).describe()
stats
```

Hypothesis:
There is a significant difference in the mean BMI between individuals with diabetes (Outcome = 1) and individuals without diabetes (Outcome = 0) in the dataset.

To test this hypothesis,I tested the assumptions of normality and equal variances.

```python
from scipy import stats

diabetes_group = df[df['Outcome'] == 1]['BMI']
non_diabetes_group = df[df['Outcome'] == 0]['BMI']

# Shapiro-Wilk test for normality
_, p_value_diabetes = stats.shapiro(diabetes_group)
_, p_value_non_diabetes = stats.shapiro(non_diabetes_group)

# Print the test results
print("Shapiro-Wilk Test - Diabetes Group:")
print("P-value:", p_value_diabetes)
print()

print("Shapiro-Wilk Test - Non-Diabetes Group:")
print("P-value:", p_value_non_diabetes)
```

```python
from scipy import stats

# Levene's test for equal variances
_, p_value_equal_var = stats.levene(diabetes_group, non_diabetes_group)

# Print the test result
print("Levene's Test:")
print("P-value:", p_value_equal_var)
```

The assumption of equal variance was met but not normality assumption. So non-parametric test (Mann_Whitney U test) was used at 5% significance level.  

```python
diabetes_group = df[df['Outcome'] == 1]['BMI']
non_diabetes_group = df[df['Outcome'] == 0]['BMI']

# Mann-Whitney U test
statistic, p_value = stats.mannwhitneyu(diabetes_group, non_diabetes_group)

# Print the test result
print("Mann-Whitney U Test:")
print("U statistic:", statistic)
print("P-value:", p_value)
```

The p-value obtained  is extremely small (9.73e-18), which is less than the significance level of 0.05. This indicates strong evidence to reject the null hypothesis and suggests that there is a significant difference in the median BMI between individuals with diabetes and individuals without diabetes. 

Therefore, based on the Mann-Whitney U test results, I can conclude that there is a statistically significant difference in the median BMI between individuals with diabetes and individuals without diabetes in the dataset. The p-value being very small suggests that the difference in BMI between the two groups is unlikely to be due to chance alone.

## Identifying and andling Outliers 

```python
df.head(5)
```

```python
# Step 1: Visualize the bmi to identify outliers
df.boxplot(column='BMI')
```

```python
# Step 2: Define the outlier criteria
# using the IQR method to identify outliers
Q1 = df['BMI'].quantile(0.25)
Q3 = df['BMI'].quantile(0.75)
IQR = Q3 - Q1
```

```python
# Step 3: Apply outlier detection method
# Identifying outliers based on the IQR method
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR
outliers = df[(df['BMI'] < lower_limit) | (df['BMI'] > upper_limit)]
outliers
```

```python
# Step 4: Handle outliers
# Replacing outliers with the mean value
bmi_mean = df['BMI'].mean()
df['BMI'] = np.where((df['BMI'] < lower_limit) | (df['BMI'] > upper_limit), bmi_mean, df['BMI'])
```

```python
# Print the cleaned dataset
print("Cleaned dataset:")
df.head()
```

By imputing the outliers with the mean value, one retains the general characteristics and trends of the data while still accounting for the presence of outliers. Imputing the outliers with the mean value allows one to retain the original sample size, which can be important for maintaining statistical power and avoiding potential biases in subsequent analyses. Imputing outliers with the mean value ensures that the imputed values are consistent with the rest of the dataset.
